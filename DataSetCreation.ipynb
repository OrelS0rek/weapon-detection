# creating the Data Set for such project

creating weapons class pictures.

```python
#downloading automatic pictures by the prompt using icrawler
from icrawler.builtin import GoogleImageCrawler
weapons_crawler = GoogleImageCrawler(
    feeder_threads=1,
    parser_threads=1,
    downloader_threads=4,
    storage={'root_dir': 'images'},
)

weapons_crawler.downloader.headers = {
    'headers': {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}}
#prompts i want the crawler to search in google browser to then download to 'images' directory
queries = ['man with handgun cctv',
           'max with knife cctv',
           'man with rifle cctv',
           'armed suspect',
           'armed terrorist',
           'cctv robbery footage',
           'cctv robbery',
           'cctv stabbing']


for query in queries:
    #scraping the images
    for year in range(2017,2025):
            weapons_crawler.crawl(keyword=query,
                            filters={'date': ((year, 1, 1), (year, 12, 31))},
                            max_num=750,
                            min_size=(200, 200),
                            max_size=None,
                            file_idx_offset='auto'
            )
```

```python
# Prompts for non-weapon CCTV images
object_crawler = GoogleImageCrawler(
    feeder_threads=1,
    parser_threads=1,
    downloader_threads=4,
    storage={'root_dir': 'objects'},
)
queries = [
    'people walking on street CCTV',
    'man carrying groceries CCTV',
    'woman holding bag CCTV',
    'office worker CCTV',
    'traffic camera footage',
    'public park CCTV',
    'elevator CCTV footage',
    'store checkout CCTV',
    'person holding scissors',
    'tools in toolbox',
    'people in a park',
    'man holding bicycle',
    'hiker with walking stick',
    'gardener holding shovel',
    'woman with suitcase',
    'artist with paintbrush',
    'plastic sword',
    'crowded subway station',

]

# Scrape images for each query
for query in queries:
    object_crawler.crawl(
        keyword=query,
        max_num=50,
        file_idx_offset='auto'
    )
```
